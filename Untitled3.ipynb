```python
import pandas as pd
df_chunk =  pd.read_csv("cleanCuiPair40det.txt", sep ="\t", chunksize = 100000)

```

```python
def chunk_preprocessing(chunk):
    chunk2 = chunk.drop_duplicates()
    chunk2 = chunk[chunk.apply(lambda x: ("alzheimer" in str(x["name1"]).lower()) or  ("alzheimer" in str(x["name2"]).lower() ),axis =1)]
    return chunk2
    
chunk_list = []  # append each chunk df here 

# Each chunk is in df format
for chunk in df_chunk:  
    # perform data filtering 
    chunk_filter = chunk_preprocessing(chunk)
    print (chunk_filter.iloc[1])
    # Once the data filtering is done, append the chunk to list
    chunk_list.append(chunk_filter)

# concat the list into dataframe 
df_concat = pd.concat(chunk_list)
```
