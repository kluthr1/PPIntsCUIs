```python
import pandas as pd
```

```python
df = pd.read_csv("../cleanCuiPair40det.txt", sep = "\t")
df
```

```python
def getSort(a, b):
    lis = [a,b]
    lis.sort()
    return lis
```

```python
df["commonC"] = df.apply(lambda x: getSort(x["cui1"] ,x["cui2"]),axis = 1)
df["commonG"] = df.apply(lambda x: getSort(x["gene1"] ,x["gene2"]) ,axis = 1)
```

```python
df["commonC"] = df.apply(lambda x: str(x["commonC"]) ,axis = 1)
df["commonG"] = df.apply(lambda x: str(x["commonG"]) ,axis = 1)
```

```python
df = df.drop_duplicates(subset = ["commonC", "commonG"])
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_json("cleanCuiPair40det.json", orient="index")
```
